```python
#Genetic Algorithm for personalized Discounts
import numpy as np
import random

# Define the problem
def profit_function(chromosome, customer_data):
    total_profit = 0
    for i in range(len(customer_data)):
        discount = chromosome[i]
        loyalty = customer_data[i]['LoyaltyScore']
        frequency = customer_data[i]['PurchaseFrequency']
        recency = customer_data[i]['Recency']

        # Simple profit function based on discount and purchase probability
        purchase_probability = (loyalty / 100) * (1 - discount) * (frequency / (recency + 1))
        if random.random() < purchase_probability:
            total_profit += (1 - discount) * customer_data[i]['BaseProfit']

    return total_profit

# Genetic Algorithm parameters
population_size = 50
generations = 100
mutation_rate = 0.1
customer_data = [
    {'LoyaltyScore': 80, 'PurchaseFrequency': 12, 'Recency': 30, 'BaseProfit': 20},
    {'LoyaltyScore': 60, 'PurchaseFrequency': 8, 'Recency': 60, 'BaseProfit': 15},
    {'LoyaltyScore': 40, 'PurchaseFrequency': 3, 'Recency': 90, 'BaseProfit': 10},
    # Add more customer data as needed
]
population = [np.random.uniform(0, 0.25, len(customer_data)) for _ in range(population_size)]
for generation in range(generations):
    # Evaluate fitness of each chromosome
    fitness = [profit_function(chromosome, customer_data) for chromosome in population]

    # Selection - pick the top 50% of the population
    selected_indices = np.argsort(fitness)[-population_size//2:]
    selected_population = [population[i] for i in selected_indices]

    # Crossover - create the next generation
    next_generation = []
    for _ in range(population_size//2):
        parent1, parent2 = random.sample(selected_population, 2)
        crossover_point = random.randint(1, len(customer_data)-1)
        child1 = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
        child2 = np.concatenate((parent2[:crossover_point], parent1[crossover_point:]))
        next_generation.extend([child1, child2])

    # Mutation - randomly mutate some of the genes
    for chromosome in next_generation:
        if random.random() < mutation_rate:
            mutation_point = random.randint(0, len(customer_data)-1)
            chromosome[mutation_point] = np.random.uniform(0, 0.25)

    population = next_generation

# Final best solution
best_solution = population[np.argmax([profit_function(chromosome, customer_data) for chromosome in population])]
print(f"Optimal Discounts: {best_solution}")
count = 0
for i in best_solution:
    i *= 100
    print("Customer {} should get {:.2f}%".format(count,i)) 
    count += 1
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
#Fuzzy Logic System for Personalized Discounts
# Define fuzzy variables
loyalty = ctrl.Antecedent(np.arange(0, 101, 1), 'loyalty')
frequency = ctrl.Antecedent(np.arange(0, 13, 1), 'frequency')
recency = ctrl.Antecedent(np.arange(0, 101, 1), 'recency')
discount = ctrl.Consequent(np.arange(0, 26, 1), 'discount')

# Define membership functions for inputs
loyalty['low'] = fuzz.trimf(loyalty.universe, [0, 0, 50])
loyalty['medium'] = fuzz.trimf(loyalty.universe, [30, 50, 70])
loyalty['high'] = fuzz.trimf(loyalty.universe, [50, 100, 100])

frequency['low'] = fuzz.trimf(frequency.universe, [0, 0, 6])
frequency['medium'] = fuzz.trimf(frequency.universe, [4, 6, 8])
frequency['high'] = fuzz.trimf(frequency.universe, [6, 12, 12])

recency['recent'] = fuzz.trimf(recency.universe, [0, 0, 50])
recency['average'] = fuzz.trimf(recency.universe, [30, 50, 70])
recency['old'] = fuzz.trimf(recency.universe, [50, 100, 100])

# Define membership functions for output
discount['low'] = fuzz.trimf(discount.universe, [0, 0, 10])
discount['medium'] = fuzz.trimf(discount.universe, [5, 12, 20])
discount['high'] = fuzz.trimf(discount.universe, [15, 25, 25])

# Define fuzzy rules
rule1 = ctrl.Rule(loyalty['high'] & frequency['high'] & recency['recent'], discount['low'])
rule2 = ctrl.Rule(loyalty['medium'] & frequency['medium'] & recency['average'], discount['medium'])
rule3 = ctrl.Rule(loyalty['low'] & frequency['low'] & recency['old'], discount['high'])

# Create control system and simulation
discount_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
discount_sim = ctrl.ControlSystemSimulation(discount_ctrl)

# Sample input for a specific customer
discount_sim.input['loyalty'] = 70
discount_sim.input['frequency'] = 8
discount_sim.input['recency'] = 40

# Compute the discount
discount_sim.compute()

# Output the recommended discount
print(f"Recommended Discount: {discount_sim.output['discount']:.2f}%")
```
